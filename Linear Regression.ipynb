# Simple Regression

```python
import pandas as pd
dataset = pd.read_csv("Company_Profit.csv")
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
li_model = LinearRegression()
li_model.fit(x_train,y_train)
```

```python
print(li_model.score(x_train,y_train))
```

```python
y_predict = li_model.predict(x_test)
y_predict
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_train,y_train,color='blue')
plt.plot(x_train,li_model.predict(x_train),color = 'red')
plt.title('Salary Vs Exp')
plt.xlabel('years of exp')
plt.ylabel('salary')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
x_serial = list(range(1, len(y_predict) + 1))
plt.scatter(x_serial, y_predict, color = 'red')
plt.scatter(x_serial, y_test, color = 'blue')
plt.title('y_pred(red) vs y_test(blue)')
plt.xlabel('Serial Number')
plt.ylabel('Profit')
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'black')
plt.plot(x_train, li_model.predict(x_train), color = 'red')
plt.title('Test Set -> Startup_Yrs_Operation Vs Profit')
plt.xlabel('Startup_Yrs_Operation')
plt.ylabel('Profit')
plt.show()
```
